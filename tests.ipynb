```python
import torch
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(0, 2, 101)
y = x.exp() - 1
z = (x+0.2)**-1
plt.plot(x, y, label='distance error')
plt.plot(x, z, label='reconstruction error')
plt.legend()
plt.axis('off')
plt.show()
```
