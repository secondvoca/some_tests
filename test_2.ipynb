```python
import torch

import matplotlib.pyplot as plt
```

```python
m = torch.distributions.MultivariateNormal(torch.zeros(2), torch.tensor([[1., 0.5], [0.5, 1]]))
s = m.sample([1000])  # normally distributed with mean=`[0,0]` and covariance_matrix=`I`
```

```python
plt.Figure(figsize=[5, 5])
plt.scatter(s[:, 0], s[:, 1])
```
